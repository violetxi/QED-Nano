```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.patches import FancyBboxPatch, Patch
```

## IMO Proof Bench

```python
# -----------------------------
# Data
# -----------------------------
categories = ["Overall", "Basic", "Advanced"]
methods = [
    "SFT",
    "Instruct",
    "GRPO",
    "Self Distillation",
    "Dense Outcome",
    "Dense Process",
]
values = {
    "SFT":                [1.4810, 2.2393, 0.7417],
    "Instruct":           [3.9234, 4.7094, 3.1441],
    "GRPO":               [3.2911, 4.1780, 2.4118],
    "Self Distillation":  [3.3924, 4.2564, 2.5500],
    "Dense Outcome":      [3.7311, 4.6891, 2.7731],
    "Dense Process":      [3.9231, 4.6261, 3.2437],
}

# Colors
colors = {
    "SFT": "#B9BDC7",
    "Instruct": "#DD8A57",
    "GRPO": "#CDA847",
    "Self Distillation": "#9FBE69",
    "Dense Outcome": "#46B58C",
    "Dense Process": "#35B8C9",
}

# -----------------------------
# Helper: rounded bars
# -----------------------------
def rounded_bar(ax, x_center, height, width, color, radius=0.04, bottom=0):
    patch = FancyBboxPatch(
        (x_center - width / 2, bottom),
        width,
        height - bottom,
        boxstyle=f"round,pad=0,rounding_size={radius}",
        linewidth=0,
        facecolor=color,
        edgecolor=color,
        mutation_aspect=1,
    )
    ax.add_patch(patch)
    return patch

# -----------------------------
# Figure
# -----------------------------
fig, ax = plt.subplots(figsize=(10, 6), dpi=150)

fig.patch.set_facecolor("#f2f2f2")
ax.set_facecolor("#f2f2f2")

x = np.arange(len(categories))
n_methods = len(methods)
bar_width = 0.12
offsets = (np.arange(n_methods) - (n_methods - 1) / 2) * bar_width

for i, method in enumerate(methods):
    for j in range(len(categories)):
        rounded_bar(
            ax=ax,
            x_center=x[j] + offsets[i],
            height=values[method][j],
            width=bar_width * 0.92,
            color=colors[method],
            radius=0.035,
        )

# -----------------------------
# Styling
# -----------------------------
ax.set_xlim(-0.55, len(categories) - 0.45)
ax.set_ylim(0, 5.2)

ax.set_xticks(x)
ax.set_xticklabels(categories, fontsize=13, fontweight="bold")
ax.set_ylabel("Average Grade", fontsize=13, fontweight="bold")
ax.set_title("IMO-ProofBench", fontsize=18, fontweight="bold", pad=14)

ax.yaxis.grid(True, color="#d9d9d9", linewidth=1)
ax.set_axisbelow(True)

ax.spines["top"].set_visible(False)
ax.spines["right"].set_visible(False)
ax.spines["left"].set_color("#666666")
ax.spines["bottom"].set_color("#666666")

ax.set_yticks(np.arange(0, 5.5, 0.5))
ax.tick_params(axis="y", labelsize=11)
ax.tick_params(axis="x", pad=10)

legend_handles = [Patch(facecolor=colors[m], edgecolor=colors[m], label=m) for m in methods]
leg = ax.legend(
    handles=legend_handles,
    title="Legend",
    loc="upper center",
    bbox_to_anchor=(0.5, -0.18),
    ncol=3,
    frameon=False,
    fontsize=11,
    title_fontsize=12,
    handlelength=1.0,
    handleheight=1.0,
    columnspacing=1.2,
    handletextpad=0.6,
)
leg._legend_box.align = "left"

plt.tight_layout(rect=[0, 0.08, 1, 1])
plt.show()
```
